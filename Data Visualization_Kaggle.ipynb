### 4. Scatter Plots

```python
# Using insurace data

insurance = pd.read_csv(r"C:\Users\TrinhQuynh\DA_Learning\Python\Data Visualizaton_Kaggle\insurance.csv")
insurance
```

<b>To create a simple scatter plot, we use the sns.scatterplot command and specify the values for:

<the horizontal x-axis (x=insurance_data['bmi']), >
    and
the vertical y-axis (y=insurance_data['charges']).

```python
# Setup
plt.figure(figsize=(14,6))

# Plot
sns.scatterplot(x=insurance['bmi'], y=insurance['charges'])
```

To double-check the strength of this relationship, you might like to add a regression line, or the line that best fits the data. We do this by changing the command to sns.regplot

```python
# Setup
plt.figure(figsize=(14,6))

# Regression plot
sns.regplot(x=insurance['bmi'], y=insurance['charges'])
```

#### Color-coded scatter plots

For instance, to understand how smoking affects the relationship between BMI and insurance costs, we can color-code the points by 'smoker', and plot the other two columns ('bmi', 'charges') on the axes.

```python
# Setup
plt.figure(figsize=(14,6))

# Plot
sns.scatterplot(x=insurance['bmi'], y=insurance['charges'], hue=insurance['smoker'])
```

#### Adding regression lines corresponding to smokers and nonsmokers.

```python
list(insurance.columns)
```

```python
# Setup
plt.figure(figsize=(14,6))

# Plot using lmplot
sns.lmplot(data=insurance, x='bmi', y='charges', hue='smoker')

plt.show()
```

### 4. Categorical scatter plot

```python
# Setup
plt.figure(figsize=(14,6))

# Plot using lmplot
sns.swarmplot(data=insurance, y='charges', x='smoker')

plt.show()
```

### 5. Histogram

```python
# Say we would like to create a histogram to see how petal length varies in iris flowers. We can do this with the sns.histplot command.
sns.histplot(iris['Petal Length (cm)'])
```

### 6. Density plots

```python
# To make a KDE plot, we use the sns.kdeplot command. 
# Setting shade=True colors the area below the curve (and data= chooses the column we would like to plot).
sns.kdeplot(iris['Petal Length (cm)'], shade=True)
```

<b> 2D KDE plots

```python
# We're not restricted to a single column when creating a KDE plot.
# We can create a two-dimensional (2D) KDE plot with the sns.jointplot command.
sns.jointplot(x=iris['Sepal Width (cm)'],y=iris['Petal Length (cm)'], kind='kde')
sns.jointplot(data=iris, x='Sepal Width (cm)', y='Petal Length (cm)', kind='kde')
```

<b> Color-coded plots

```python
list(iris.columns)
```

```python
# Histogram of Petal Lengths, by Species
sns.histplot(data=iris, x= 'Petal Length (cm)', hue='Species')
plt.title('Histogram of Petal Lengths, by Species')
```

```python
# KDE plots for each species
sns.kdeplot(data=iris, x= 'Petal Length (cm)', hue='Species', shade=True)
plt.title('KDE plots for each species')
```
